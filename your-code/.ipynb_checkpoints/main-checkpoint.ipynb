```python
import pandas as pd
import matplotlib_inline
import chart_studio.plotly as pl
import cufflinks as cf
from ipywidgets import interact

cf.go_offline()
```

## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
df=pd.read_excel('../data/Online Retail.xlsx')
```

```python
df
```

```python
data=df[df.Country!='United Kingdom']
data=data.groupby('Country').sum()

data=data[['Quantity', 'Revenue']]
data.iplot(kind='bar', xTitle='Country', title='Cantidad Total y Beneficio')
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
data=df[(df.InvoiceDate >= '2011-01-01') & (df.InvoiceDate < '2011-06-01') & (df.Country == 'France')]

data = data[['Quantity', 'Revenue']]

data.iplot(kind = 'line', xTitle='Country', title='Cantidad Total y Beneficio')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
data=df[df.Description=='PARTY BUNTING']

data=data.groupby('Country').mean()
data=data[['Quantity', 'UnitPrice']]

data.reset_index(inplace=True)

data.iplot(x='Quantity', y='UnitPrice', categories='Country',
           xTitle='Average quantity', yTitle='Average unit price', title='Average by Country');
```

## 5. Create an interactive side-by-side bar chart showing the revenue by country listed below (bars) for each of the products listed below.

```python
products =['JUMBO BAG RED RETROSPOT', 
          'CREAM HANGING HEART T-LIGHT HOLDER',
          'REGENCY CAKESTAND 3 TIER']

countries =['EIRE', 'Germany', 'France']

data=df[df.Description.isin(products)]
data=data[data.Country.isin(countries)]



data=data.pivot_table(values='Revenue', columns='Description',
                      index='Country', aggfunc='sum')

data.reset_index(inplace=True)


data.iplot(kind='bar', x='Country', title='Revenue by Country',
          xTitle='Country', yTitle='Revenue')
```
